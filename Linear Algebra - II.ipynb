### If anyone of a matrix's eigen value is zero, then product of the eigen values must be zero and determinant must also be zero.

# Singular Value Decomposition (SVD)## A = $UDV^T$ 
$$



```python
import numpy as np
```

```python
A = np.array([[-1,2],[3,-2],[5,7]])
A
```

```python
U,D,VT = np.linalg.svd(A) 
```

```python
U
```

```python
D
```

```python
VT
```

```python
np.diag(D)
```

### D must have the same dimensions as A for $UDV^T$ matrix multiplication to be possible

```python
D = np.concatenate((np.diag(D),[[0,0]]),axis = 0)
```

```python
D
```

```python
np.dot(U,np.dot(D,VT))
```

### Image Compression via SVD

```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('dolly.jpg')
_=plt.imshow(img)
```

### Convert Image to grayscale

```python
img_gray = img.convert('LA')
img_gray
```

### Convert data into numpy matrix

```python
import numpy as np
```

```python
# Convert the image to grayscale
imggray = img.convert('L')
```

```python
imgmat = np.array(list(imggray.getdata(band=0)), float)
imgmat.shape = (imggray.size[1],imggray.size[0])
imgmat = np.matrix(imgmat)
_=plt.imshow(imgmat,cmap='gray')
```

### Calculate SVD of the Image

```python
U,sigma,V = np.linalg.svd(imgmat)
```

```python
U
```

```python
sigma
```

```python
V
```

```python
reconstimg = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
_ = plt.imshow(reconstimg, cmap='gray')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming U, sigma, and V are already computed from the SVD of A
# U, sigma, V = np.linalg.svd(A, full_matrices=False)

for i in [2, 4, 8, 16, 32, 64,128]:
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```
